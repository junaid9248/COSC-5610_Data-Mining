# Task 2: Exploring NumPy

```python
import numpy as np
import os
import skimage.io
import matplotlib.pyplot as plt
```

```python
bush=np.zeros((250,250,3),dtype=np.float64) 
count=0
for (root,dirs,files) in os.walk('/Users/ther3/Marquette/Spring 2025/COSC 5610 - Data Mining/codes/Discussion 1/George_W_Bush_dataset', topdown=True):
# need to replace the above directory with your image directory
    for name in files:
        bush=bush+skimage.io.imread(os.path.join(root, name))
        count+=1
```

```python
print("processed images:", count)
bush=(bush/count).astype(int) 

#imshow requires values either floats in [0, 1] or integers in [0, 255]. Thus, bush /= count*255 also works
skimage.io.imshow(bush)
# check the basic information of the array
print(bush.ndim, bush.shape, bush.dtype)
```
